### Recorrer todas las columnas y eliminar aquellas que tengan "--" como valor

```python
for col in all_columns:
    current_column = df[col].values

    all__ = lambda el: el != "--"

    any_true = any(map(all__, current_column))

    if any_true == False:
        print(f"Columna eliminada: {col}")
        df.drop(col, axis="columns", inplace=True)
```
