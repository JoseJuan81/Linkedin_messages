### Filtrar n칰meros telef칩nico de Per칰

```python
filt_more_than_2 = df["Phone 1 - Value"].str.len() > 2
filt_51 = df["Phone 1 - Value"].str.startswith("+51")
filt_9 = df["Phone 1 - Value"].str.startswith("9")
filt_01 = df["Phone 1 - Value"].str.startswith("01")
filt_PE = filt_more_than_2 & (filt_51 | filt_9 | filt_01)
```

```python
show_columns = ["Name", "Notes", "E-mail 1 - Value", "Phone 1 - Value", "Phone 2 - Value", "Organization 1 - Name", "Organization 1 - Title"]
peru = df.loc[filt_PE, show_columns]
```

### Negar el filtro u obtener resultados opuestos

```python
df.loc[~filt_PE, show_columns].head()
```
